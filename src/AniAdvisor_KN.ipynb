```python
field_descriptions = {
    "user_id": "non identifiable randomly generated user id",
    "anime_id": "MyAnimelist ID of the anime",
    "score": "score between 1 to 10 given by the user (0 if not assigned)",
    "MAL_ID": "MyAnimelist ID of the anime",
    "Name": "full name of the anime",
    "Score": "average score of the anime from all users",
    "Genres": "comma separated list of genres",
    "English name": "full name in english of the anime",
    "Type": "TV, movie, OVA, etc.",
    "Episodes": "number of chapters/episodes",
    "Aired": "broadcast date",
    "Producers": "comma separated list of producers",
    "Studios": "comma separated list of studios",
    "Duration": "duration per episode",
    "Rating": "age rating",
    "Ranked": "position based on the score",
    "Popularity": "position based on the number of users who added the anime",
    "Members": "number of community members in this anime's group",
    "Favorites": "number of users who marked the anime as favorites",
    "Score-10": "number of users who scored 10",
    "Score-9": "number of users who scored 9",
    "Score-8": "number of users who scored 8",
    "Score-7": "number of users who scored 7",
    "Score-6": "number of users who scored 6",
    "Score-5": "number of users who scored 5",
    "Score-4": "number of users who scored 4",
    "Score-3": "number of users who scored 3",
    "Score-2": "number of users who scored 2",
    "Score-1": "number of users who scored 1"
}

# Ensure we only include keys that are present in the DataFrame.
field_descriptions = {k: v for k, v in field_descriptions.items() if k in anime_df.columns}

# Function to combine all remaining fields with their descriptions.
def combine_all_fields(row):
    parts = []
    for field, description in field_descriptions.items():
        parts.append(f"{field} ({description}): {row[field]}")
    return ". ".join(parts)

# Create a new column 'combined_text' that holds the comprehensive description.
anime_df['combined_text'] = anime_df.apply(combine_all_fields, axis=1)

# Optionally, save the dataset with the combined text.
anime_df.to_csv('/content/drive/MyDrive/Aniadvisor/anime_combined.csv', index=False)
print("Dataset with combined text saved as 'anime_reduced_combined.csv'.")

# Preview a few rows of the combined text.
anime_df[['combined_text']].head()
```

```python
# Load the sentence transformer model
model = SentenceTransformer('all-mpnet-base-v2')

# Generate embeddings for the combined text field; progress bar helps visualize the process.
embeddings = model.encode(anime_df['combined_text'].tolist(), show_progress_bar=True)

# Add embeddings to your DataFrame; note that each embedding is a numpy array.
anime_df['embedding'] = embeddings.tolist()

# Optionally, save the DataFrame with embeddings for later use
anime_df.to_csv('anime_with_embeddings.csv', index=False)
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def recommend_anime(user_input, anime_df, model, top_n=5):
    # Generate embedding for the user query; the model will use GPU if available.
    query_embedding = model.encode([user_input])[0]

    # Stack all anime embeddings into a matrix for similarity computation
    anime_embeddings = np.vstack(anime_df['embedding'].values)

    # Compute cosine similarity between the query embedding and each anime embedding
    similarity_scores = cosine_similarity([query_embedding], anime_embeddings)[0]

    anime_df['similarity'] = similarity_scores

    # Sort by similarity score in descending order and return top_n recommendations
    recommendations = anime_df.sort_values(by='similarity', ascending=False).head(top_n)
    return recommendations[['MAL_ID', 'Name_synopsis', 'Genres_synopsis', 'sypnopsis', 'similarity']]

# Example usage:
user_preference = ("I want to watch long running animes similar to bleach, naruto and onepiece, but not them")
recommended_anime = recommend_anime(user_preference, anime_df, model, top_n=25)
print(recommended_anime)
```

```python
import pandas as pd
import itertools

# Function to compute Jaccard similarity between two lists
def jaccard_similarity(list1, list2):
    set1, set2 = set(list1), set(list2)
    union = set1.union(set2)
    if not union:
        return 0
    return len(set1.intersection(set2)) / len(union)

# Assume anime_df is your DataFrame and already has the "combined_text" column.
# Also, we assume "Genres_synopsis" contains comma-separated genres.
# For efficiency, we may want to sample a subset if your dataset is very large.
sample_df = anime_df.sample(n=500, random_state=42)  # Adjust the sample size as needed

pairs = []
# Iterate over all unique pairs (you could also use more intelligent pairing if desired)
for idx1, idx2 in itertools.combinations(sample_df.index, 2):
    row1 = sample_df.loc[idx1]
    row2 = sample_df.loc[idx2]

    # Process genres into lists (strip whitespace)
    genres1 = [g.strip() for g in str(row1['Genres_synopsis']).split(',')]
    genres2 = [g.strip() for g in str(row2['Genres_synopsis']).split(',')]

    # Compute a similarity score using Jaccard similarity
    score = jaccard_similarity(genres1, genres2)

    # Create a training pair using the combined text of each anime.
    pairs.append({
        "sentence1": row1["combined_text"],
        "sentence2": row2["combined_text"],
        "score": score
    })

# Create a DataFrame from the pairs
pairs_df = pd.DataFrame(pairs)

# Save the pairs to a CSV file
pairs_df.to_csv("/content/drive/MyDrive/Aniadvisor/anime_similarity.csv", index=False)
print("Generated anime_similarity.csv with {} pairs.".format(len(pairs_df)))
```
